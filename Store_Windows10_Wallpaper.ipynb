```python
# Method 1, using two near identical copy-loops
import getpass
import os, shutil
from pathlib import Path
from PIL import Image


## Specific for you as a user, adjust freely ##
destination_folder_desktop = 'D://Wallpapers' # e.g. D://Wallpapers//Desktop
destination_folder_phone = 'D://Wallpapers//Cell Phone' # e.g. D://Wallpapers//Phone


## Initiate variables
windows_user = getpass.getuser() # if an error occurs with the user, please hardcode windows user here
source_folder = ('C://Users//' + windows_user + '//AppData//Local//Packages//Microsoft.Windows.ContentDeliveryManager_cw5n1h2txyewy//LocalState//Assets')
if destination_folder_desktop == '':
    destination_folder_desktop = 'C://Users//' + windows_user + '//Desktop'
if destination_folder_phone == '':
    destination_folder_phone = 'C://Users//' + windows_user + '//Desktop'
copy_to_desktop = []
copy_to_phone = []
counter_new_files_desktop = 0
counter_new_files_phone = 0
counter_renames = 0


## Index all new files in source location
for new_file in os.listdir(source_folder):
    # Index images to phone or desktop (and check whether it is an hd wallpaper):
    img = Image.open(os.path.join(source_folder, new_file))
    if img.size == (1920, 1080):
        copy_to_desktop.append(new_file) # Desktop Wallpaper
    elif img.size == (1080, 1920):
        copy_to_phone.append(new_file) # Phone wallpaper


## Copy files
for desktop_file in copy_to_desktop:
    # Check first if file already exists, if not continue
    if not os.path.exists(os.path.join(destination_folder_desktop, desktop_file + '.jpg')):
        shutil.copy(os.path.join(source_folder, desktop_file),
                    os.path.join(destination_folder_desktop, desktop_file + '.jpg'))
        counter_new_files_desktop += 1

for phone_file in copy_to_phone:
    # Check first if file already exists, if not continue
    if not os.path.exists(os.path.join(destination_folder_phone, phone_file + '.jpg')):
        shutil.copy(os.path.join(source_folder, phone_file),
                    os.path.join(destination_folder_phone, phone_file + '.jpg'))
        counter_new_files_phone += 1


## Rename file extensions, if empty make jpg. Sometimes neccessary for Windows 10
counter_renames = 0
for file in os.listdir(destination_folder_desktop):
    if os.path.isfile(os.path.join(destination_folder_desktop, file)):
        if Path(file).suffix == '':
            os.rename(os.path.join(destination_folder_desktop, file),
                        os.path.join(destination_folder_desktop, file + '.jpg'))
            counter_renames += 1


## Finished
print('Finished!\nCopied desktop images \t = ' + str(counter_new_files_desktop) \
      + '\nCopied phone images \t = ' + str(counter_new_files_phone) \
      + '\nRenamed files \t\t = ' + str(counter_renames))
```

```python
# Method 2, using a function for copying
import getpass
import os, shutil
from pathlib import Path
from PIL import Image


## Specific for you as a user, adjust freely ##
destination_folder_desktop = 'D://Wallpapers' # e.g. D://Wallpapers//Desktop
destination_folder_phone = 'D://Wallpapers//Cell Phone' # e.g. D://Wallpapers//Phone


## Code start
def copy_files(source_folder, files_to_copy, destination_folder):
    '''
        Description:    Copies image files according to a list
        Input:          List with image files, full filepath of source and destination folder
        Remark:         .jpg is added
    '''
    counter = 0
    for image_file_to_copy in files_to_copy:
        # Check first if file already exists, if not continue
        if not os.path.exists(os.path.join(destination_folder, image_file_to_copy + '.jpg')):
            shutil.copy(os.path.join(source_folder, image_file_to_copy), 
                        os.path.join(destination_folder, image_file_to_copy + '.jpg'))
            counter += 1
    return counter


## Initiate variables
windows_user = getpass.getuser() # if an error occurs with the user, please hardcode windows user here
source_folder = ('C://Users//' + windows_user + '//AppData//Local//Packages//Microsoft.Windows.ContentDeliveryManager_cw5n1h2txyewy//LocalState//Assets')
if destination_folder_desktop == '':
    destination_folder_desktop = 'C://Users//' + windows_user + '//Desktop'
if destination_folder_phone == '':
    destination_folder_phone = 'C://Users//' + windows_user + '//Desktop'
copy_to_desktop = []
copy_to_phone = []
counter_new_files_desktop = 0
counter_new_files_phone = 0
counter_renames = 0


## Index all new files in source location
for new_file in os.listdir(source_folder):
    # Index images to phone or desktop (and check whether it is an hd wallpaper):
    img = Image.open(os.path.join(source_folder, new_file))
    if img.size == (1920, 1080):
        copy_to_desktop.append(new_file) # Desktop Wallpaper
    elif img.size == (1080, 1920):
        copy_to_phone.append(new_file) # Phone wallpaper


## Copy files
counter_new_files_desktop = copy_files(source_folder, copy_to_desktop, destination_folder_desktop)
counter_new_files_phone = copy_files(source_folder, copy_to_phone, destination_folder_phone)


## Rename file extensions, if empty make jpg. Sometimes neccessary for Windows 10
counter_renames = 0
for file in os.listdir(destination_folder_desktop):
    if os.path.isfile(os.path.join(destination_folder_desktop, file)):
        if Path(file).suffix == '':
            os.rename(os.path.join(destination_folder_desktop, file),
                        os.path.join(destination_folder_desktop, file + '.jpg'))
            counter_renames += 1


## Finished
print('Finished!\nCopied desktop images \t = ' + str(counter_new_files_desktop) \
      + '\nCopied phone images \t = ' + str(counter_new_files_phone) \
      + '\nRenamed files \t\t = ' + str(counter_renames))
```
